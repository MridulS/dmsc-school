```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
true = 12
```

```python
mu = 10.4; sigma = 1.6
```

```python
n = norm(mu, sigma)
```

```python
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
```

```python
plt.subplots(figsize=(5,3))
plt.plot(x, n.pdf(x))
plt.axvline(true, color='k')
plt.ylim(0, None)
plt.xlim(x.min(), x.max())
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.tight_layout()
plt.savefig('./images/normal.png', dpi=600)
```
